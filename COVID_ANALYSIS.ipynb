# COVID_19 ANALYSIS

### IMPORTING THE DATASET 

```python
import pandas as pd 
import numpy as np 
c_d = pd.read_csv("covid19countryinfo.csv")
df=pd.DataFrame(c_d)
```

### CLEANING DATASET(removing nullvalues)

```python
cols = ['country','pop']
df[cols] = df[cols].fillna(0)
# df.groupby(by='country')
df
```

```python
numeric_cols = [
    'pop', 'cases', 'newcases', 'deaths', 'newdeaths',
    'recovered', 'critical', 'tests',
    'casediv1m', 'deathdiv1m',
    'density', 'medianage', 'urbanpop'
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

### TOP 10 COUNTRIES WITH HIGH TOTAL_CASES

```python
df['totalcases'] = df['totalcases'].astype(str)
df['totalcases'] = df['totalcases'].str.replace(',', '', regex=True).astype(float)
df_sorted = df.sort_values(by='totalcases', ascending=False)
print(df_sorted[['totalcases','country']].head(10))
```

```python
%matplotlib inline 
import matplotlib.pyplot  as plt 
fig,ax= plt.subplots()
x=df_sorted['totalcases'].head(10)
y=df_sorted['country'].head(10)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.xlabel("COUNTRIES")
plt.ylabel("TOTAL_CASES")
plt.title("TOP 10 COUNTRIES WITH HIGH TOTAL_CASES")
ax.bar(y,x);
```

### TOTAL CASES VS TOTAL DEATHS (COUNTRY-WISE)

```python
df['totalcases']=df['totalcases'].astype(float)
df['deaths']=df['deaths'].astype(str)
df['deaths'] = df['deaths'].str.replace(',', '', regex=True).astype(float)
df_agg = df.groupby('country', as_index=False)[['totalcases', 'deaths']].sum()
x = df_agg['totalcases']
y = df_agg['deaths']
plt.xlabel("TOTAL CASES")
plt.ylabel("TOTAL DEATHS")
plt.title("TOTAL CASES VS TOTAL DEATHS")
plt.scatter(x, y);
```

### INCREASE IN CASES

```python
df['firstcase'] = pd.to_datetime(df['firstcase'])
df['firstcase']=df['firstcase'].astype(str)
df['totalcases']=df['totalcases'].astype(str)
df['totalcases'] = df['totalcases'].str.replace(',', '').astype(float)
df_sorted = df.sort_values('firstcase')
plt.tight_layout()
plt.plot(df_sorted['firstcase'], df_sorted['totalcases'])
plt.xticks(rotation=45)
ax = plt.gca()
ticks = ax.get_xticks()[::3]
ax.set_xticks(ticks)
plt.tight_layout()
plt.xlabel("DATES");
plt.ylabel("TOTAL CASES");
```
